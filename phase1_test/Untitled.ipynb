```python
import pandas as pd
from fuzzywuzzy import fuzz

names = ['test_name', 'build', 'failure']
df1 = pd.read_csv('result.csv', sep='|', names=names)
tests = df1.test_name.unique()
# print(df1['failure'])
for test in tests:
    b = df1.where(df1.test_name==test)
    print(f'################# {test} ##########################')
    # print(f'{b.failure.dropna()}')
    x = list(b.failure.dropna())
    str1 = x[0]
    for i in x:
        str2 = i
        ratio_result = fuzz.ratio(str1, str2)
        print(f'STRING1: {str1}\nSTRING2: {str2}\nRATIO:{ratio_result}\n{"-"*50}')

```
